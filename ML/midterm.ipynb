### 문제 1) P 자동차 회사의 차종 (name), 마력 (horse power), 연비 (efficiency)의 data를 이용하여 선형회귀 모델을 구한다. 이 모델에서 마력이 특징값 (입력 데이터), 연비가 출력값 (target value) 이다. (30점)

```python
import pandas as pd

df = pd.DataFrame({
    'name' : ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'horse power' : [130, 230, 180,  300, 210, 220, 165],
    'efficiency': [15.3, 9.2, 10.1, 6.1, 11.1, 12.5, 13.1] })

vehicle_df = df.set_index('name')
print(vehicle_df)
```

#### (a) 선형회귀 분석을 통해서 선형회귀 모델의 절편과 계수를 구하라. 그리고 이 선형회귀 모델이 입력 마력 값에 대해 연비를 예측하는데 얼마나 적합한지 예측점수를 출력하라.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# 선형회귀 모델의 학습을 위한 code를 작성하세요.

# X: 특징값 (입력 데이터)
X = vehicle_df['horse power'].to_numpy()
X = X[:, np.newaxis]

# y: 출력값 (target value)
y = vehicle_df['efficiency'].to_numpy()

print(X)

lin_model1 = LinearRegression()
lin_model1.fit(X, y)
```

```python
print('계수 :', lin_model1.coef_)
print('절편 :', lin_model1.intercept_)
print('예측 점수 :', lin_model1.score(X, y))
```

#### (b) 위의 선형 회귀 모델을 바탕으로 270 마력의 신형엔진을 가진 자동차를 개발하려 한다. 선형 회귀 모델을 이용하여 이 자동차의 연비를 구하라. 출력은 소수점 둘째 자리까지 출력하라.

```python
# 관련 code를 작성하세요.
hp = 270
y_pred = lin_model1.predict(pd.DataFrame([[hp]]))
print(hp, "마력 차량의 예상 연비 :", np.round(y_pred, 2))
```

### 문제 2) 다음은 P 자동차 회사의 차종 (name)과 마력 (horse power), 뿐만아니라 자동차의 총중량 (weight)을 추가하였다. 이 data를 이용하여 선형회귀 모델을 구한다. 이 모델에서 마력, 총중량이 특징값 (입력 데이터)이고 연비가 출력값 (target value) 이다. (30점)

```python
import pandas as pd

df = pd.DataFrame({
    'name' : ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'horse power' : [130, 230, 180,  300, 210, 220, 165],
    'weight': [1920, 2610, 2215, 2910, 2411, 2305, 2117],
    'efficiency': [15.3, 9.2, 10.1, 6.1, 11.1, 12.5, 13.1] })

vehicle_df = df.set_index('name')
print(vehicle_df)
```

#### (a) 선형회귀 분석을 통해서 선형회귀 모델의 절편과 계수를 구하라. 그리고 이 선형회귀 모델의 예측점수를 출력하라.

```python
from sklearn.linear_model import LinearRegression

# 선형회귀 모델의 학습을 위한 code를 작성하세요.

# X: 특징값 (입력 데이터)
X = vehicle_df[['horse power', 'weight']]

# y: 출력값 (target value)
y = vehicle_df['efficiency']

print(X)

lin_model2 = LinearRegression()
lin_model2.fit(X, y)
```

```python
print('계수 :', lin_model2.coef_)
print('절편 :', lin_model2.intercept_)
print('예측 점수 :', lin_model2.score(X, y))
```

#### (b) 위의 선형 회귀 모델을 바탕으로 270마력의 신형엔진을 가진 총중량 2,500kg의 자동차를 개발하려 한다. 선형 회귀 모델을 이용하여 이 자동차의 연비를 구하라. 출력은 소수점 둘째 자리까지 출력하라.

```python
# 관련 code를 작성하세요.
hp = 270
wt = 2500
y_pred = lin_model2.predict(pd.DataFrame([[hp, wt]], columns=['horse power', 'weight']))
print(hp, "마력,", wt, "kg 차량의 예상 연비 :", np.round(y_pred, 2))
```

### 문제 3) 어느 동물병원에서 치료를 받은 개의 종류와 그 data ([길이, 높이]) 이다. 이 data를 이용하여 k-NN 분류기를 학습한다. (40점)

```python
## 닥스훈트 데이터
d =  [[75, 24], [77, 29], [83, 19], [81, 32], [73, 21], \
      [99, 22], [72, 19], [83, 34]]
### 사모예드 데이터
s = [[76, 55], [78, 58], [82, 53], [88, 54], [76, 61], \
     [83, 52], [81, 57], [89, 64]]
### 말티즈 데이터
m = [[35, 23], [39, 26], [38, 19], [41, 30], [30, 21], \
     [57, 24], [41, 28], [35, 20]]

print('닥스훈트(0) :', d)
print('사모예드(1) :', s)
print('말티즈(2) :', m)
```

### 개의 길이, 높이 데이터 A, B, C, D에 대하여 각각 n_neighbor를 3, 5, 7로 하여 분류하고 그 분류 결과를 출력하라.

<pre>
A : 길이 58, 높이 55
B : 길이 40, 높이 26
C : 길이 80, 높이 39
D : 길이 75, 높이 35
</pre>

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

dog_classes = {0:'닥스훈트', 1:'사모예드', 2:'말티즈'}

d_label = np.zeros(len(d))
s_label = np.ones(len(s))
m_label = np.full(len(m), 2)  # 수정

dogs = np.concatenate((d, s, m))
labels = np.concatenate((d_label, s_label, m_label))

# 분류기 함수 (show_classifier_result)를 위한 code
def show_classifier_result(data, label):
    for n in [3, 5, 7]:
        # K-NN 분류기 학습 및 예측 (y_pred)을 위한 code를 작성하세요.
        knn = KNeighborsClassifier(n_neighbors=n)
        knn.fit(dogs, labels)
        y_pred = knn.predict(data)
        print(label, data, ': n_neighbors가', n,'일때 :', dog_classes[y_pred[0]])

print('A 데이터 분류결과')
A = [[58, 55]]
show_classifier_result(A, 'A')

B = [[40, 26]]
print()
print('B 데이터 분류결과')
show_classifier_result(B, 'B')

C = [[80, 39]]
print()
print('C 데이터 분류결과')
show_classifier_result(C, 'C')

D = [[75, 35]]
print()
print('D 데이터 분류결과')
show_classifier_result(D, 'D')
```
